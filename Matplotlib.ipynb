```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.linspace(0,5,11)
x
```

```python
y=x**2
y
```

```python
plt.plot(x,y,'g')


plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'g')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#00
fig = plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.set_xlabel('X_Axis')
axes.set_ylabel('Y_Axis')
axes.set_title('square')

```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes1.plot(x,y)
axes1.set_title('Large')
axes2 = fig.add_axes([0.2,0.2,0.6,0.6])
axes2.plot(y,x)
axes2.set_title('small')

```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,0.8,0.8])
ax.plot(x,x*2,label='x_square')
ax.plot(x,x*3,label='x_cube')
ax.legend(loc=0)






```
